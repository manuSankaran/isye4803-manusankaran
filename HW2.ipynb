With this given txt file, could you compute (for ALL VALUES), the column means, a correlation, and a simple lm() linear fit in R? Use ROE ~ ROEt

Follow the same steps for Python, keeping a similar structure as well.

```python
dat <- read.table("/Users/manu/Downloads/roe01.txt", header = TRUE)

colMeans(dat)

cor(dat)

cor(dat$ROE, dat$ROEt)

fit <- lm(ROE ~ ROEt, data = dat)

summary(fit)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

dat = pd.read_table("/Users/manu/Downloads/roe01.txt", sep=r"\s+")

dat.mean()

dat.corr()

dat["ROE"].corr(dat["ROEt"])

X = sm.add_constant(dat["ROEt"])
y = dat["ROE"]

fit = sm.OLS(y, X).fit()

fit.summary()
```
